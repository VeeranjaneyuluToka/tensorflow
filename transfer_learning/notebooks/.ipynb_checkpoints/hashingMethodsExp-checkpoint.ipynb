```python
#extract query frame FVs
inputPath = '../../../data/movie_titles/testUncompressed/3D6C53DB-054A-4935-9872-AB2C6093C29F/5.jpg'
modelType = 'ENB5'
enFeatExt = featureExtractor(1000, modelType)
ENModel = enFeatExt.pretrainedModel()
img = enFeatExt.loadPreprocessData(inputPath, ENModel)
enpreds = ENModel.predict(img)
enpreds.shape
```

```python
qFVs = enpreds.squeeze()
lsh.query(qFVs, distance_func='cosine')
```

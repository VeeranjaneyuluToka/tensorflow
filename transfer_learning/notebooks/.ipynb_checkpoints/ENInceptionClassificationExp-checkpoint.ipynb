```python
import setGPU

import matplotlib.pyplot as plt
import numpy as np
import sys
import time
import cv2 as cv
import os

sys.path.append('..')

import tensorflow as tf
import keras.applications.imagenet_utils as imgutils

from extractEmbeddings import featureExtractor

#from efficientnet.tfkeras import EfficientNetB7
#from efficientnet.tfkeras import center_crop_and_resize, preprocess_input
```

```python
inputPath = '../../../data/movie_titles/testUncompressed/3D6C53DB-054A-4935-9872-AB2C6093C29F/5.jpg'
```

```python
def displayImage(img):
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
img = cv.imread(inputPath)
rgbImg = cv.cvtColor(img, cv.COLOR_BGR2RGB)
displayImage(rgbImg)
```

```python
modelType = 'InceptionV3'
featExt = featureExtractor(1000, modelType)
loadStartTime = time.time()
model = featExt.pretrainedModel(False)
print("model: {}, loading time: {}".format(modelType, time.time()-loadStartTime))
```

```python
img = featExt.loadPreprocessData(inputPath, model)
preds = model.predict(img)
imgutils.decode_predictions(preds)
```

```python
modelType = 'ENB7'
enFeatExt = featureExtractor(1000, modelType)
loadStartTime = time.time()
ENModel = enFeatExt.pretrainedModel(False)
print("model: {}, loading time: {}".format(modelType, time.time()-loadStartTime))
```

```python
img = enFeatExt.loadPreprocessData(inputPath, ENModel)
enpreds = ENModel.predict(img)
imgutils.decode_predictions(enpreds)
```
